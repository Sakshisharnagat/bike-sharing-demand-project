70-30 splitting model performance

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# Data for 70-30 split
data_7030 = {
    "Model": ["Linear Regression", "Decision Tree", "Random Forest", "KNN", "Gradient Boosting", "SVM"],
    "MAE": [636.381358, 575.84, 433.179318, 525.2454, 460.430095, 1666.511529],
    "MSE": [751126.231064, 793596.3, 429523.628898, 539686.0076, 423643.77255, 3922760.66645],
    "RMSE": [866.67539, 890.84, 655.38052, 734.63324, 650.87923, 1980.5960],
    "R2_Score": [0.8112257766, 0.8005, 0.892051, 0.86436, 0.89352, 0.0141256]
}

df_7030 = pd.DataFrame(data_7030)

# Set seaborn style
sns.set(style="whitegrid")

# Highlight best model: highest R2 + lowest RMSE
best_model = df_7030.loc[df_7030['R2_Score'].idxmax(), 'Model']

# Plotting
fig, ax1 = plt.subplots(figsize=(12,6))

# Bar width
bar_width = 0.2
x = np.arange(len(df_7030['Model']))

# Plot MAE, MSE, RMSE as separate bars
ax1.bar(x - bar_width, df_7030['MAE'], width=bar_width, color='orange', label='MAE')
ax1.bar(x, df_7030['MSE'], width=bar_width, color='blue', label='MSE')
ax1.bar(x + bar_width, df_7030['RMSE'], width=bar_width, color='red', label='RMSE')

# Left Y-axis for errors
ax1.set_ylabel('Error Values', fontsize=12)
ax1.set_xticks(x)
ax1.set_xticklabels(df_7030['Model'], rotation=30, ha='right')
ax1.set_title('ML Model Performance (70-30 Split)', fontsize=16)

# Second Y-axis for R² Score
ax2 = ax1.twinx()
ax2.plot(x, df_7030['R2_Score'], color='green', marker='o', linewidth=2, label='R² Score')
ax2.set_ylabel('R² Score', color='green', fontsize=12)
ax2.tick_params(axis='y', labelcolor='green')

# Highlight best model
for i, model in enumerate(df_7030['Model']):
    if model == best_model:
        ax2.annotate('Best Model', xy=(i, df_7030['R2_Score'][i]+0.02),
                     xytext=(i, df_7030['R2_Score'][i]+0.05),
                     arrowprops=dict(facecolor='green', shrink=0.05),
                     ha='center', fontsize=10, color='green')

# Legends
fig.legend(loc="upper right", bbox_to_anchor=(0.9,0.9))
plt.tight_layout()
plt.show()
```
